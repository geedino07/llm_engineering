## Day 1 exercise - summarize a webpage




A Jupyter Notebook that fetches webpage content via HTTP request, extracts readable text, and generates an AI-powered summary.

```python
# imports dependencies
from IPython.display import Markdown, display
from openai import OpenAI
from bs4 import BeautifulSoup
import requests


headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}
```

```python
#instantiate openai object
client = OpenAI()

#set the system and user prompts 
system_prompt = """
You are a snarky assistant that analyzes the contents of an email,
and provides a short, snarky, humorous summary with a perfect subject line,
ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""

user_prompt = """
Summarize the website with a perfect subject line.
Please let your response be short, snarky and humorous.
"""

def summarize(url):
    website = fetch_website_contents(url)

    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": website + "\n\n" + user_prompt}
    ]
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=messages
    )
    return response.choices[0].message.content



def fetch_website_contents(url):
    """
    Return the title and contents of the website at the given url;
    truncate to 2,000 characters as a sensible limit
    """
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    title = soup.title.string if soup.title else "No title found"
    if soup.body:
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        text = soup.body.get_text(separator="\n", strip=True)
    else:
        text = ""
    return (title + "\n\n" + text)[:2_000]

# retireve and display result
result = summarize("https://geedino07.github.io/cybersecurity-scam-alert/")
display(Markdown(result))
```
